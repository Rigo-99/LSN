<p style="color:#180090; font-size: 0.9cm; padding-top: 1em; font-weight:bold;">Esercizio 2</p>

<div style="color:#333333; font-size: 0.43cm; padding: 1em; background-color:#F5F5F5">
Parallelizzare il codice dell'algoritmo genetico per risolvere il problema del commesso viaggiatore con 4 esecuzioni in parallelo. Ognuno dei 4 nodi deve eseguire indipendentemente i calcoli dell'algoritmo genetico, ma ogni $N_{migr}$ generazioni i vari nodi si devono scambiare i propri migliori individui.
    
Affrontare così il problema delle $32$ città in un quadrato.
</div>

<div style="font-size: 0.43cm;">
I quattro processi (continenti) vengono eseguiti indipendentemente fino allo scambio degli individui (atleti), ogni $N_{migr}$ generazioni (olimpiade).

Per architettare lo scambio di atleti, si è seguita la logica dell'equità tra i continenti. Sarebbe inutile che in un continente finiscano tutti quelli peggiori, mentre in un altro tutti i migliori.
Pertanto, senza neppure utilizzare tempo di calcolo per centralizzare la ridistribuzione casuale degli atleti, si è pensato di organizzare le olimpiadi in questo modo:<br>
ogni continente comunica a tutti gli altri (tramite $\texttt{Bcast}$) tutti i propri atleti. Così facendo, ogni continente ha l'elenco completo degli atleti di tutto il mondo. A questo punto ogni continente aggiorna il proprio elenco di atleti: il continente $k$ prende come proprio nuovo $i$-esimo atleta, il vecchio $i$-esimo atleta dal continente $(k+i) \text{ mod } 4$.
    
Seguendo questa semplice regola (deterministica) di ridistribuzione degli atleti, ogni continente riesce a calcolare chi "prendersi" e chi invece viene preso da altri senza bisogno di comunicare nuovamente con gli altri continenti. Inoltre, in questo modo, si garantisce maggiore equità tra i continenti.
    


</div>

```python
best0, media0, sigma0 = np.loadtxt("Esercitazione_10_2/results/best_0.dat", usecols=(0,1,2), unpack='true')
best1, media1, sigma1 = np.loadtxt("Esercitazione_10_2/results/best_1.dat", usecols=(0,1,2), unpack='true')
best2, media2, sigma2 = np.loadtxt("Esercitazione_10_2/results/best_2.dat", usecols=(0,1,2), unpack='true')
best3, media3, sigma3 = np.loadtxt("Esercitazione_10_2/results/best_3.dat", usecols=(0,1,2), unpack='true')
x = np.arange(len(best0))
plt.plot(x,best0,label="continente 0")
plt.plot(x,best1,label="continente 1")
plt.plot(x,best2,label="continente 2")
plt.plot(x,best3,label="continente 3")
plt.xlabel('generazioni')
plt.ylabel('lunghezza percorso')
plt.grid(True)
plt.title('Percorso minimo')
plt.legend()
plt.show()

x = np.arange(len(best0))
plt.plot(x,best0,label="continente 0")
plt.plot(x,best1,label="continente 1")
plt.plot(x,best2,label="continente 2")
plt.plot(x,best3,label="continente 3")
plt.xlabel('generazioni')
plt.ylabel('lunghezza percorso')
plt.xlim(0,400)
plt.grid(True)
plt.title('Percorso minimo [zoom]')
plt.legend()
plt.show()
```

<div style="font-size: 0.43cm;">
Tutti i continenti giungono ad una situazione stabile dopo circa $300$ generazioni. Nel caso del singolo processore (Esercitazione 9), ce ne erano volute circa $600$.
    
Si vede bene dal grafico quanto siano importanti le olimpiadi (in questo caso ogni $80$ generazioni): alla generazione $80$, $160$ e $240$ si vede bene come i continenti meno performanti vengono immediatamente resi competitivi dallo scambio con gli altri.
</div>
