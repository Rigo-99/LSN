```python
best0, media0, sigma0 = np.loadtxt("results/best_0.dat", usecols=(0,1,2), unpack='true')
best1, media1, sigma1 = np.loadtxt("results/best_1.dat", usecols=(0,1,2), unpack='true')
best2, media2, sigma2 = np.loadtxt("results/best_2.dat", usecols=(0,1,2), unpack='true')
best3, media3, sigma3 = np.loadtxt("results/best_3.dat", usecols=(0,1,2), unpack='true')
x = np.arange(len(best0))
plt.plot(x,best0,label="continente 0")
plt.plot(x,best1,label="continente 1")
plt.plot(x,best2,label="continente 2")
plt.plot(x,best3,label="continente 3")
plt.xlabel('generazioni')
plt.ylabel('lunghezza percorso')
plt.grid(True)
plt.title('Percorso minimo')
plt.legend()
plt.show()
```

```python
file = open ( 'map.dat' , 'r')
mappa = []
mappa = [ line.split() for line in file]

G = nx.Graph()

for city in mappa:
    G.add_node(city[0], pos=[float(city[1]),float(city[2])])

pos = nx.get_node_attributes(G, 'pos')
plt.figure(3,figsize=(5,5)) 
nx.draw(G, pos=pos, node_size=100)
plt.show()
```

```python
file = open ( 'results/path_1.dat' , 'r')
path = []
path = [ line.split() for line in file]

G.add_edges_from(path)

pos = nx.get_node_attributes(G, 'pos')
plt.figure(3,figsize=(5,5)) 
nx.draw(G, pos=pos, node_size=100, edge_color = 'red')
plt.show()

file = open ( 'results/path_2.dat' , 'r')
path = []
path = [ line.split() for line in file]

G.add_edges_from(path)

pos = nx.get_node_attributes(G, 'pos')
plt.figure(3,figsize=(5,5)) 
nx.draw(G, pos=pos, node_size=100, edge_color = 'red')
plt.show()
```
